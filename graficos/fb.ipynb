```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def listador(res):
    f = open(res, 'r')
    c = 0
    l_alguna_vacia = []
    l_mismo = []
    l_distinto = []
    p_alguna_vacia = []
    p_mismo = []
    p_distinto = []
    for line in f:
        if 'Tiempo' in line:
            tiempo = int(line.split()[-2])
            if c < 40:
                l_alguna_vacia.append(tiempo)
            elif c < 80:
                l_mismo.append(tiempo)
            else:
                l_distinto.append(tiempo)
            c += 1
        elif 'distancia' in line:
            p1 = str(line.split()[6][1:-1])
            p2 = str(line.split()[8][1:-1])
            l = [len(p1), len(p2)]
            if c < 20:
                p_alguna_vacia.append(l[1])
            elif c < 40:
                p_alguna_vacia.append(l[0])
            elif c < 80:
                p_mismo.append(l[0])
            else:
                p_distinto.append(abs(l[0]-l[1]))
    f.close()
    return l_alguna_vacia, p_alguna_vacia, l_mismo, p_mismo, l_distinto, p_distinto


      
```

```python
res = '../generadores/salida_bf.txt'
l_alguna_vacia, p_alguna_vacia, l_mismo, p_mismo, l_distinto, p_distinto= listador(res)

res2 = '../generadores/salida_pd.txt'
l_alguna_vacia2, p_alguna_vacia2, l_mismo2, p_mismo2, l_distinto2, p_distinto2 = listador(res2)
```

```python
def y(x):
    return (4**x)

def y2(x):
    return(x**2)

def y3(x):
    return (2**x)
    

x = range(0, 10)
x = np.linspace(0, 10, 100)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(p_alguna_vacia, l_alguna_vacia, label = "Fuerza Bruta", color = 'hotpink', alpha=0.5)
plt.scatter(p_alguna_vacia2, l_alguna_vacia2, label = "Programación Dinámica", color='gold', alpha=0.5)
plt.plot(x, [y3(i) for i in x], label="2^x", color="blue", alpha = 0.5)
plt.plot(x, [y2(i) for i in x], label="x^2", color="darkorchid", alpha = 0.5)

plt.ylim(0, 1000000) 
plt.xlabel('Largo de la cadena no vacía')
plt.ylabel('Tiempo de ejecución (nanosegundos)')
plt.title('Tiempo de ejecución con alguna cadena vacía')
plt.grid()
plt.legend()
plt.show()
```

```python


plt.figure(figsize=(10, 6))

plt.scatter(p_mismo, l_mismo, label = "Fuerza Bruta", color = 'hotpink', alpha=0.5)
plt.scatter(p_mismo2, l_mismo2, label = "Programación Dinámica", color='gold', alpha=0.5)
plt.plot(x, [y(i) for i in x], label="4^x", color="blue", alpha = 0.5)
plt.plot(x, [y2(i) for i in x], label="x^2", color="darkorchid", alpha = 0.5)

plt.ylim(0, 1000000) 
plt.xlabel('Largo de las cadenas')
plt.ylabel('Tiempo de ejecución (nanosegundos)')
plt.title('Tiempo de ejecución con ambos largos iguales')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(p_distinto, l_distinto, label = "Fuerza Bruta", color = 'hotpink', alpha=0.5)
plt.scatter(p_distinto2, l_distinto2, label = "Programación Dinámica", color='gold', alpha=0.5)
plt.plot(x, [y(i) for i in x], label="4^x", color="blue", alpha = 0.5)
plt.plot(x, [y2(i) for i in x], label="x^2", color="darkorchid", alpha = 0.5)

plt.ylim(0, 1000000) 
plt.xlabel('Diferencia de largos')
plt.ylabel('Tiempo de ejecución (nanosegundos)')
plt.title('Tiempo de ejecución con largos distintos')
plt.grid()
plt.legend()
plt.show()
```
